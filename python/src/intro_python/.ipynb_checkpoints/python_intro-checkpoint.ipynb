````python
import numpy as np
import matplotlib.pyplot as plt

# Simulamos los datos de las 20 compa침칤as durante los 칰ltimos 4 a침os
np.random.seed(42)  # For reproducibility
companies = [f"Compa침칤a {i+1}" for i in range(20)]
years = np.array([2020, 2021, 2022, 2023])

# Generamos precios de las acciones en un rango de 50 a 500 para el a침o 2020, y le damos una variaci칩n para los dem치s a침os
stock_prices = np.random.uniform(50, 500, (20, 4)) + np.random.normal(0, 20, (20, 4))

# Simulamos los cambios en el precio de las acciones tomando en cuenta eventos hipot칠ticos (como la necedad gringa de invadir pa칤ses)
event_impact = np.array([0, -30, 50, -20])  # Impacto por a침o 2021 ca칤da, 2022 alza, etc.
stock_prices += event_impact

# Normalizamos los datos
normalized_prices = (stock_prices - np.min(stock_prices)) / (np.max(stock_prices) - np.min(stock_prices))

# Generamos los cambios en los precios de las acciones
plt.figure(figsize=(14, 8))
for i, company in enumerate(companies):
    plt.plot(years, normalized_prices[i, :], label=company, alpha=0.7)

# Resaltamos los a침os en los que hubo un impacto militar
for year, impact in zip(years, event_impact):
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)
    plt.text(year, 1.05, f"{impact:+} Impacto", ha='center', fontsize=10, color='red')

# Configuramos la gr치fica
plt.title("Cambios en el precio de las acciones de las 20 compa침ias del Fortune 500 (2020-2023)")
plt.xlabel("A침o")
plt.ylabel("Precio normalizado de la acci칩n")
plt.legend(loc='upper left', fontsize=8, ncol=2)
plt.grid(alpha=0.3)
plt.tight_layout()

# Show the plot
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Simulamos los datos de las 20 compa침칤as durante los 칰ltimos 4 a침os
np.random.seed(42)  # For reproducibility
companies = [f"Compa침칤a {i+1}" for i in range(20)]
years = np.array([2020, 2021, 2022, 2023])

# Generamos precios de las acciones en un rango de 50 a 500 para el a침o 2020, y le damos una variaci칩n para los dem치s a침os
stock_prices = np.random.uniform(50, 500, (20, 4)) + np.random.normal(0, 20, (20, 4))

# Simulamos los cambios en el precio de las acciones tomando en cuenta eventos hipot칠ticos (como la necedad gringa de invadir pa칤ses)
event_impact = np.array([0, -30, 50, -20])  # Impacto por a침o 2021 ca칤da, 2022 alza, etc.
stock_prices += event_impact

# Normalizamos los datos
normalized_prices = (stock_prices - np.min(stock_prices)) / (np.max(stock_prices) - np.min(stock_prices))

# Preparamos la gr치fica a animar
fig, ax = plt.subplots(figsize=(14, 8))
lines = []
for _ in companies:
    line, = ax.plot([], [], alpha=0.7)
    lines.append(line)

# Resaltamos los a침os en los que hubo un impacto militar
for year, impact in zip(years, event_impact):
    ax.axvline(x=year, color='gray', linestyle='--', alpha=0.5)
    ax.text(year, 1.05, f"{impact:+} Impact", ha='center', fontsize=10, color='red')

ax.set_xlim(years[0], years[-1])
ax.set_ylim(0, 1.1)
ax.set_title("Cambios en el precio de las acciones de las 20 compa침ias del Fortune 500 (2020-2023)")
ax.set_xlabel("A침o")
ax.set_ylabel("Precio normalizado de la acci칩n")
ax.grid(alpha=0.3)

# Funci칩n de inicializaci칩n
def init():
    for line in lines:
        line.set_data([], [])
    return lines

# Animation function
def animate(frame):
    year_idx = frame
    for i, line in enumerate(lines):
        line.set_data(years[:year_idx + 1], normalized_prices[i, :year_idx + 1])
    return lines

# Creamos la animaci칩n
ani = FuncAnimation(fig, animate, frames=len(years), init_func=init, blit=True, interval=1000, repeat=False)

# Mostramos la animaci칩n (descomentar si corremos por consola
# plt.show()

# Mostramos la animaci칩n (descomentar si corremos desde un notebook)
from IPython.display import HTML
HTML(ani.to_jshtml())
````
