# Introducción a Python

## Instalación de python

### Linux

Python viene instalado por defecto en todas las distribuciones de linux, si se desea instalar una versión especifica se aconseja utilizar una herramienta como `pipenv` o `virtualenv`. `pip` no viene incluído por defecto así que debes instalarlo por tu cuenta.

### Windows

Python se puede [instalar](https://learn.microsoft.com/en-us/windows/python/beginners) desde la `Microsoft Store`, sólo hay que buscar "Python" y darle clic en `instalar`, estas versiones de python incluyen `pip` por defecto.

### MacOs

La manera más fácil de obtener una versión específica de python es utilizando las [imágenes oficiales de instalación](https://www.python.org/downloads/macos/), si la versión instalada no incluye `pip` éste se puede instalar utilizando:

```bash
curl https://bootstrap.pypa.io/get-pip.py -o get-pip.py
```

## Instalando dependencias en python

La popularidad de Python es también su maldición, debido a que la gente busca cooperar con más personas se necesita un sistema de administración de paquetes (y todos los problemas que eso lleva), el administrador por defecto en Python se llama `venv` y viene incluído en cualquier instalación moderna de Python (de la versión `3.3` en adelante).

Sin embargo la comunidad a creado un conjunto de administradores de ambientes/paquetes que se utilizan para satisfacer necesidades específicas de sus usuarios.

### Administradores de ambientes
#### Pyenv

Es una [herramienta](https://github.com/pyenv/pyenv) que nos permite cambiar la versión de python de manera rápida (tarea moral: ¿Por qué quisiéramos hacer eso?), las características principales de la herramienta son: cambio automático de versiones, instalación fácil y aislamiento de verisones.

La instalación de dicha herramienta es trivial en linux:

```bash
curl https://pyenv.run | bash
```

#### Anaconda

Es la [distribución](https://docs.anaconda.com/anaconda/) de Python más popular, su propósito es tener un ambiente universal para realizar cualquier tipo de análisis de Ciencias de Datos utilizando Python como lenguaje de programación, la distribución incluye:

- `conda` Un administrador de ambientes que nos permite hacer instalaciones locales a un proyecto.
- `Anaconda Navigator` Una aplicación de escritorio que te permite hacer todo lo que haces a través de la línea de comandos.
- Más de 300 bibliotecas de python para que puedas trabajar después de instalar `Anaconda` (`scikit`, `numpy`, `pandas`, etc).
- Un repositorio con bibliotecas (llamadas paquetes) relacionadas a machine learning y ciencia de datos.
  
#### Miniconda

Es una versión de peso ligero de `Anaconda`, ésta sólo instala lo básico necesario para funcionar (no incluye los paquetes de Ciencias de Datos), reutiliza muchos componentes de `conda`. Puedes leer más sobre la distribución [aquí](https://docs.anaconda.com/miniconda/).

#### Pipenv

Es un [administrador de versiones de python](https://pipenv.pypa.io/en/latest/) cuyo objetivo principal es brindar ergonomía a los desarrolladores ya que permite "fijar" las depedencias de un proyecto a través de un archivo `Pipfile.lock`, no es necesario utilizar un archivo `requirements.txt` donde se especifican hashes, la instalación de dependencias a través de `pip` son manejadas de manera automática, por debajo hace uso de `virtualenv` para lograr una experiencia de desarrollo placentera, la instalación es fácil:

```bash
pip install --user pipenv
```

#### Virtualenv

Es una [herramienta](https://virtualenv.pypa.io/en/latest/index.html) que te permite crear ambientes aislados de python, un subconjunto de esta herramienta ha sido integrado de manera éstandar en Python (desde la versión 3.3), las diferencias entre `virtualenv` y `venv` son:

- `venv` es más lento (pero no se necesita instalar nada).
- `venv` no soporta extensiones.
- `venv` no puede crear ambientes virtuales para una versión aleatorea de python (o identificar la versión de un proyecto aleatoreo).
- `venv` no se puede actualizar utilizando `pip`.

## Proyecto de ejemplo

A manera de ejemplo crearemos un proyecto simple que hará uso de `matplotlib` y `numpy` para graficar algunos datos ficticios de las 20 compañías más exitosas del `Fortune 500`, para ello haremos uso de `pipenv` con el fin de instalar una versión específica de Python y declarar las dependencias del sistema.

### Creando el proyecto

Antes de iniciar con algo necesitamos saber qué es lo que hará nuestro sistema, para ello necesitamos "nombrar" el proyecto y su "alcance", en particular vámos a listar lo que hará el proyecto:

- Obtener las 20 compañías más redituables del Fortune 500.
- Determinar el tamaño de la ventana de tiempo (supongamos 4 años)
- Obtener el comportamiento del precio de las acciones de cada compañía.
- Graficar el comportamiento promedio de las acciones para dichas compañías
- Animar la gráfica porque podemos 😉

El proyecto se hará en 2 etapas: la primer etapa nos ayudará a comprender la forma en la que utilizamos `pipenv` para crear un proyecto y define el esqueleto funcional del proyecto, la segunda etapa obtendrá datos reales basándose en el código que creamos en la primer etapa.

Lo primero que necesitamos hacer es "levantar un ambiente" en una carpeta específica ejecutando los comandos:

```bash
mkdir fortune500 && cd fortune500 # Creamos una carpeta específica para el proyecto
pipenv shell
```

Esto creará de manera automática el ambiente, por lo que podemos instalar las dependencias del proyecto con:

```bash
pipenv install numpy matplotlib
```

Como consecuencia se generará un archivo `Pipfile` que lista todas las dependencias del proyecto, el contenido de este archivo debe ser similar a lo siguiente:

```
[[source]]
url = "https://pypi.org/simple"
verify_ssl = true
name = "pypi"

[packages]
numpy = "*"
matplotlib = "*"

[dev-packages]

[requires]
python_version = "3.12"
```

Si deseamos instalar una versión específica podemos hacerlo de la siguiente manera `pipenv install numpy==x.y.z`, donde `x.y.z` es la versión específica que buscamos. ¿Qué pasa si encontramos una biblioteca súper interesante en Github o en alguna otra forja pero dicha bibiloteca todavía no se encuentra dada de alta en el repositorio estándar de bibilotecas de pyton (`Pypi`)? Pues con `pipenv` podemos indicar la ruta de dicha biblioteca y la herramienta se encarga de instalar de manera automática todo:

```bash
pipenv install -e git+https://github.com/soy_un_programador/biblioteca_bonita.git
```

Cuando creamos un proyecto es imperativo crear un conjunto de pruebas para verificar que no estamos programando barbaridades, la biblioteca predilecta para hacer pruebas en python se llama `pytest` (pero no es la única), lo "malo" de una suite de pruebas es que, si nuestro código necesita correr en un servidor de "producción" toda la suite de pruebas no es necesaria después de que nuestro código ha sido desplegado en un ambiente productivo, para evitar instalar cosas innecesarias podemos indicar que ciertas dependencias sólo serán utilizadas mientras estamos desarrollando, para marcar una biblioteca como "modo desarrollo" utilizamos la bandera `--dev`.

```bash
pipenv install pytest --dev
```

De esta forma el archivo `Pipfile` contendrá una sección llamada `[dev-packages]` que no se instalarán cuando creemos una versión de "producción" de nuestro código. Una vez finalizada la instalación de dependencias (y después de asegurarnos que nuestro código funciona correctamente) **debemos** fijar las versiones de las bibliotecas que utilizamos para garantizar una mejor compatibilidad:

```bash
pipenv lock
```

Esto creará el archivo `Pipfile.lock`, dicho archivo contiene el cálculo de dependencias necesarias del proyecto utilizando las versiones específicas que usaste al momento de programarlo, minimizando compatibilidades, como consejo nunca deberías de editar manualmente dicho archivo. ¿Y cómo instalo la versión de "producción"? Ejecutando el comando:

```bash
pipenv install --ignore-pipfile
```

Esto hace que `pipenv` utilice el archivo `Pipfile.lock` para obtener las versiones específicas necesarias para correr el proyecto. Como extra, si quieres ver el árbol de dependencias de tu proyecto puedes utilizar el comando `pipenv graph`.

#### Abriendo dependencias

En ocasiones necesitamos ver cómo es que una biblioteca implementa cierto método, para explorar esto `pipenv` ofrece el siguiente comando:

```
pipenv open numpy
```

Esto abrirá el proyecto `numpy` utilizando el valor de la variable de entorno `EDITOR`.

#### Variables de entorno

Podemos crear un archivo `.env` en el que declaramos variables de entorno que pueden ser utilizadas por nuestros proyectos, `pipenv` se encarga de leer este archivo y pasarle las variables a nuestro programa, la lectura del archivo `.env` se realiza al momento de ejecutar el comando `pipenv shell`, el archivo `.env` tiene la siguiente estructura:

```
VARIABLE1=valor
VARIABLE2=valor
.
.
.
VARIABLEN=valor
```

#### ¿Cómo cierro mi ambiente virtual?

Para salir del ambiente virtual simplemente ejecutamos el comando `exit` 😋

#### Código base para el análisis

Continuando con la parte de programación en este punto crearemos el código que, dados los datos de las 20 compañías más redituables del fortune 500 crea una gráfica para visualizar el cambio en el precio de sus acciones, para ello iniciaremos importando las dependencias del proyecto:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Es aconsejable usar el `as` ya que de esta forma cualquier función que utilicemos de la biblioteca contendrá el prefijo indicado (`np` para `numpy` y `plt` para `matplotlib`), lo primero que haremos es **simular** que tenemos los datos, para ello generaremos de manera aleatoria la información sobre las compañías:

```python
companies = [f"Compañía {i+1}" for i in range(20)]
years = np.array([2020, 2021, 2022, 2023])

# Generamos precios de las acciones en un rango de 50 a 500 para el año 2020, y le damos una variación para los demás años
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Simulamos los datos de las 20 compañías durante los últimos 4 años
np.random.seed(42)  # La semilla nos permite reproducir los datos aleatoreos
companies = [f"Compañía {i+1}" for i in range(20)]
years = np.array([2020, 2021, 2022, 2023])

# Generamos precios de las acciones en un rango de 50 a 500 para el año 2020, y le damos una variación para los demás años
stock_prices = np.random.uniform(50, 500, (20, 4)) + np.random.normal(0, 20, (20, 4))
````

Esto nos devuelve un arreglo multidimensional de dimension $20$ columnas cada una de ellas teniendo $4$ filas (una por año). Para simular el impacto de un evento externo en el precio de las acciones generamos un arreglo de $4$ elementos y "sumamos" el valor del impacto a cada una de las entradas por año:

```python
# Simulamos los cambios en el precio de las acciones tomando en cuenta eventos hipotéticos (como la necedad gringa de invadir países)
event_impact = np.array([0, -30, 50, -20])  # Impacto por año 2021 caída, 2022 alza, etc.
stock_prices += event_impact
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Simulamos los datos de las 20 compañías durante los últimos 4 años
np.random.seed(42)  # For reproducibility
companies = [f"Compañía {i+1}" for i in range(20)]
years = np.array([2020, 2021, 2022, 2023])

# Generamos precios de las acciones en un rango de 50 a 500 para el año 2020, y le damos una variación para los demás años
stock_prices = np.random.uniform(50, 500, (20, 4)) + np.random.normal(0, 20, (20, 4))

# Simulamos los cambios en el precio de las acciones tomando en cuenta eventos hipotéticos (como la necedad gringa de invadir países)
event_impact = np.array([0, -30, 50, -20])  # Impacto por año 2021 caída, 2022 alza, etc.
stock_prices += event_impact

# Normalizamos los datos
normalized_prices = (stock_prices - np.min(stock_prices)) / (np.max(stock_prices) - np.min(stock_prices))

# Generamos los cambios en los precios de las acciones
plt.figure(figsize=(14, 8))
for i, company in enumerate(companies):
    plt.plot(years, normalized_prices[i, :], label=company, alpha=0.7)

# Resaltamos los años en los que hubo un impacto militar
for year, impact in zip(years, event_impact):
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)
    plt.text(year, 1.05, f"{impact:+} Impacto", ha='center', fontsize=10, color='red')

# Configuramos la gráfica
plt.title("Cambios en el precio de las acciones de las 20 compañias del Fortune 500 (2020-2023)")
plt.xlabel("Año")
plt.ylabel("Precio normalizado de la acción")
plt.legend(loc='upper left', fontsize=8, ncol=2)
plt.grid(alpha=0.3)
plt.tight_layout()

# Show the plot
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Simulamos los datos de las 20 compañías durante los últimos 4 años
np.random.seed(42)  # For reproducibility
companies = [f"Compañía {i+1}" for i in range(20)]
years = np.array([2020, 2021, 2022, 2023])

# Generamos precios de las acciones en un rango de 50 a 500 para el año 2020, y le damos una variación para los demás años
stock_prices = np.random.uniform(50, 500, (20, 4)) + np.random.normal(0, 20, (20, 4))

# Simulamos los cambios en el precio de las acciones tomando en cuenta eventos hipotéticos (como la necedad gringa de invadir países)
event_impact = np.array([0, -30, 50, -20])  # Impacto por año 2021 caída, 2022 alza, etc.
stock_prices += event_impact

# Normalizamos los datos
normalized_prices = (stock_prices - np.min(stock_prices)) / (np.max(stock_prices) - np.min(stock_prices))

# Preparamos la gráfica a animar
fig, ax = plt.subplots(figsize=(14, 8))
lines = []
for _ in companies:
    line, = ax.plot([], [], alpha=0.7)
    lines.append(line)

# Resaltamos los años en los que hubo un impacto militar
for year, impact in zip(years, event_impact):
    ax.axvline(x=year, color='gray', linestyle='--', alpha=0.5)
    ax.text(year, 1.05, f"{impact:+} Impact", ha='center', fontsize=10, color='red')

ax.set_xlim(years[0], years[-1])
ax.set_ylim(0, 1.1)
ax.set_title("Cambios en el precio de las acciones de las 20 compañias del Fortune 500 (2020-2023)")
ax.set_xlabel("Año")
ax.set_ylabel("Precio normalizado de la acción")
ax.grid(alpha=0.3)

# Función de inicialización
def init():
    for line in lines:
        line.set_data([], [])
    return lines

# Animation function
def animate(frame):
    year_idx = frame
    for i, line in enumerate(lines):
        line.set_data(years[:year_idx + 1], normalized_prices[i, :year_idx + 1])
    return lines

# Creamos la animación
ani = FuncAnimation(fig, animate, frames=len(years), init_func=init, blit=True, interval=1000, repeat=False)

# Mostramos la animación (descomentar si corremos por consola
# plt.show()

# Mostramos la animación (descomentar si corremos desde un notebook)
from IPython.display import HTML
HTML(ani.to_jshtml())
````

#### Descargando información real

Para esta parte necesitamos la biblioteca `yfinance` para descargar los datos desde Yahoo Finance:

````python
import yfinance as yf

# En teoría tomamos los tickets de las 20 compañías mas redituables del Fortune 500
tickers = ["AAPL", "MSFT", "AMZN", "GOOGL", "TSLA"]  # Reemplazar por el ticker correcto

# Descargamos el precio de las acciones de los últimos 4 años para los tickers
data = yf.download(tickers, start="2020-01-01", end="2023-12-31", group_by="ticker")

# Mostramos las primeras filas
for ticker in tickers:
    print(f"\n{ticker} Precio de las acciones:")
    print(data[ticker].head())
````
