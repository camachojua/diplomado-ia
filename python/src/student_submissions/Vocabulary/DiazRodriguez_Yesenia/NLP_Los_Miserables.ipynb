```python
import fitz  # PyMuPDF
```

```python
ruta_pdf = "/content/drive/My Drive/Los-miserables.pdf"
```

```python
def pdf_a_texto(ruta_pdf):
    doc = fitz.open(ruta_pdf)
    texto = ""
    for pagina in doc:
        texto += pagina.get_text("text") + "\n"
    return texto
texto = pdf_a_texto(ruta_pdf)
print(texto[:1000])  # Muestra los primeros 1000 caracteres
```

```python
import re
from unidecode import unidecode
```

```python
def limpiar_texto(texto):
    texto = texto.lower()
    texto = unidecode(texto)  # Elimina acentos
    texto = re.sub(r'[^\w\s]', '', texto)  # Quita puntuaci√≥n
    texto = re.sub(r'\d+', '', texto)  # Elimina n√∫meros
    texto = re.sub(r'\s+', ' ', texto).strip()  # Quita espacios extra
    return texto

texto_limpio = limpiar_texto(texto)
print(texto_limpio[:1000])  # Ver los primeros 1000 caracteres
```

```python
import nltk
```

```python
# Download the 'punkt_tab' resource
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
```

```python
from collections import Counter # Import the Counter class
```

```python
# Tokenizar el texto limpio
tokens = word_tokenize(texto_limpio)

print("Primeros 10 tokens:", tokens[:10])
```

```python
# Construir el vocabulario con frecuencia de palabras
frecuencia_palabras = Counter(tokens)
```

```python
# Las 100 palabras m√°s y menos frecuentes
mas_frecuentes = frecuencia_palabras.most_common(100)
menos_frecuentes = frecuencia_palabras.most_common()[:-101:-1]
```

```python
total_palabras = len(tokens)
palabras_unicas = len(frecuencia_palabras)
longitud_promedio = sum(len(p) for p in tokens) / total_palabras
```

```python
print(f'üìå N√∫mero total de palabras: {total_palabras}')
print(f'üìå N√∫mero de palabras √∫nicas: {palabras_unicas}')
print(f'üìå Longitud promedio de las palabras: {longitud_promedio:.2f}\n')
```
